### Hashcat Wordlist Attack Command Structure

To perform a wordlist attack with Hashcat, use the following format:

```bash
./hashcat-6.2.6/hashcat.bin -a 0 -m [HASH_TYPE] [YOUR_HASH_OR_HASHFILE] [YOUR_WORDLIST_PATH]
```

*   `-a 0` specifies the **Straight** attack mode (wordlist attack).
*   `-m [HASH_TYPE]` defines the hash type. You must replace `[HASH_TYPE]` with the numeric ID of your hash. You can find these IDs in the `hashcat -h` output (e.g., `0` for MD5, `1000` for NTLM).
*   `[YOUR_HASH_OR_HASHFILE]` should be replaced with the actual hash you want to crack, or the path to a file containing one or more hashes.
*   `[YOUR_WORDLIST_PATH]` is the path to your wordlist file. For example, if you downloaded `rockyou.txt`, its path would be `/content/wordlists/rockyou.txt`.

**Example:** If you have an MD5 hash `d41d8cd98f00b204e9800998ecf8427e` and want to use `rockyou.txt`:

```bash
./hashcat-6.2.6/hashcat.bin -a 0 -m 0 d41d8cd98f00b204e9800998ecf8427e /content/wordlists/rockyou.txt
```
